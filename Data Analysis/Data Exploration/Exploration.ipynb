#### Image Visualization:
Display a random subset of images to visually inspect the quality, size, and content. This can help identify any data quality issues or variations.
1. Sample Display:
- Display a subset of images from your dataset to get a general sense of the data's content, quality (such as corruption, noise, or artifacts) and variability.

```python

plt.figure()
for i, image_file in enumerate(training_data):
    image_file = image_file.astype(float)
    # image_path = os.path.join(image_dir, image_file)
    # img = cv2.imread(image_dir)
    # img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(3, 3, i + 1)
    plt.imshow(image_file)
    plt.title(f"fig {str(i)}.jpg")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

4. Histogram of Pixel Intensities:
Plot histograms of pixel intensities to understand the distribution of brightness values in the images. This can help you identify potential lighting variations or anomalies.



```python
import numpy as np
plt.figure(figsize=(12, 8))
for i, image_file in enumerate(training_data):
    image_file = image_file.astype(float)
    plt.subplot(5,2, i + 1)
    plt.hist(np.ravel(image_file), bins=256, range=(0, 256), color='black', alpha=0.9)
    plt.title(f"fig {str(i)}.jpg")
plt.tight_layout()
plt.show()
```

5. Channel Visualization:
For color images (e.g., RGB), visualize individual color channels (Red, Green, Blue) separately to observe how features are represented in each channel.


```python
import cv2

image_dir = '/Users/amberm/PycharmProjects/ProteinSub/data/img'
image_files = os.listdir(image_dir)
img_list = ["red", "green", "blue", "yellow"]
image_list = [i+".png" for i in img_list]

plt.figure(figsize=(12, 12))
for i, img_file in enumerate(image_list):
    image_path = os.path.join(image_dir, img_file)
    img = cv2.imread(image_path)
    plt.subplot(2, 2, i + 1)
    plt.imshow(img)
    plt.title(img_file)
    plt.axis('off')
plt.tight_layout()
plt.show()
```
