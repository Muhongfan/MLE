# Basic example with simple input
## Line

```python
import matplotlib.pyplot as plt

x1 = [10,20,30]
y1 = [20,40,10]
# line 2 points
x2 = [10,20,30]
y2 = [40,10,30]

```

```python
# Set a title 
plt.title('Two or more lines with different widths, colors, linestyle, markers')
# Display the figure.
plt.plot(x1,y1)
plt.plot(x2,y2)

# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')

plt.title('Basic')
plt.show()
```

## Basic property 
### line property

```python

# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title 
plt.title('Two or more lines with different widths, colors, linestyle, markers')
# Display the figure.
plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-width-3', marker='o', markerfacecolor='yellow', markersize=12)
plt.plot(x2,y2, color='red', linewidth = 5,  label = 'line2-width-5',linestyle='dotted')
# show a legend on the plot
plt.show()
```

### Figure property

```python

# Set the x axis label of the current axis.
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title 
plt.title('Display the current axis limits values and set new axis values')
# Display the figure.
plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-width-3', marker='o', markerfacecolor='yellow', markersize=12)
plt.plot(x2,y2, color='red', linewidth = 5,  label = 'line2-width-5',linestyle='dotted')


print(plt.axis()) 
# set new axes limits
# Limit of x axis 5 to 35
# Limit of y axis 5 to 45
plt.axis([5, 35, 5, 45]) 


# show a legend on the plot
plt.show()
```

## Scatter - two methods

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = [10,20,30,15,25,35]
y1 = [10,20,40,60,80,100]


x2 = [5,3,40,60,80,100]
y2 = [10,20,30,15,25,35]

# 1. direct plot 
plt.plot(x1, y1, 'r*', x2, y2, 'b^')

# 2. scatter plot 
plt.scatter(x1, y1, color = 'r', marker = '*')
plt.scatter(x2, y2, color = 'b', maker = '^')



plt.title('scatter with star markers and color')
plt.show()
```

# Different decoration
## 1. show a legend

```python
## with legend()
import matplotlib.pyplot as plt
# line 1 points
x1 = [10,20,30]
y1 = [20,40,10]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")
# line 2 points
x2 = [10,20,30]
y2 = [40,10,30]
# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Two or more lines on same plot with suitable legends ')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

2. with axis

```python
import datetime as DT
from matplotlib import pyplot as plt
from matplotlib.dates import date2num

data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]

fig = plt.figure()

graph = fig.add_subplot(111)

# Plot the data as a red line with round markers
graph.plot(x,y,'r-o')

# Set the xtick locations
graph.set_xticks(x)
# Set the xtick labels
graph.set_xticklabels(
        [date.strftime("%Y-%m-%d") for (date, value) in data]
        )

# naming the x axis
plt.xlabel('Date')
# naming the y axis
plt.ylabel('Closing Value')
 # giving a title  
plt.title('With grid and xticks') 



# Customize the grid
plt.grid(linestyle='-', linewidth='0.5', color='blue')
plt.show()

```

## Bar 

```python
import matplotlib.pyplot as plt

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color=(0.4, 0.6, 0.8, 1.0))
plt.xticks(x_pos, x)
plt.show()
```

1. vertical bar 

```python
plt.barh(x_pos, popularity, color='green')

plt.yticks(x_pos, x)
plt.show()
```

2. label on the bar

```python

import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]

fig, ax = plt.subplots()
rects1 = ax.bar(x_pos, popularity, color='b')

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Label on each bar")
plt.xticks(x_pos, x)

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%f' % float(height),
        ha='center', va='bottom')
autolabel(rects1)

plt.show()


```

## Pie 

```python
import matplotlib.pyplot as plt
# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)  
# Plot
plt.pie(popuratity, explode=explode, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()

```
